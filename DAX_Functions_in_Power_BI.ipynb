Q1. Create a calculated column TotalPrice using Quantity and Unit Price.

- Answer 1

TotalPrice = 'Table'[Quantity] * 'Table'[UnitPrice]

Q2. Create a calculated column DiscountedPrice.

- Answer 2

DiscountedPrice = 'Table'[SalesAmount] * (1 - 'Table'[Discount])

Q3. Create a measure Average Order Value (AOV).

- Answer 3

AOV = AVERAGE('Table'[SalesAmount])

Q4. Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order Date

- Answer 4

YTD Sales = TOTALYTD(SUM('Table'[SalesAmount]), 'Table'[OrderDate])

Q5. Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling product ranked first.

- Answer 5

Product Rank = RANKX(ALL('Table'[ProductID]), SUM('Table'[SalesAmount]), , DESC)

Q6. Create a calculated column classifying orders as “High Value” if sales > 2000, else “Low Value”.

- Answer 6

Order Category = IF('Table'[SalesAmount] > 2000, "High Value", "Low Value")

Q7. Create a measure to calculate total sales from the Online channel only.

- Answer 7

Online Sales = CALCULATE(SUM('Table'[SalesAmount]), 'Table'[SalesChannel] = "Online")

Q8. Create a measure called "Date Difference" that calculates the number of days between the latest  Order Date and today's date using the DATEDIFF function.

- Answer 8

Date Difference = DATEDIFF( MAX('Table'[OrderDate]), TODAY(), DAY)
